# JSON

```python
data = { "president": 
{ "name": "Zaphod Beeblebrox",
 "species": "Betelgeusian", 
"male": 1 
} }

import json 
type(data)#Data Type is Dictionary
```

```python
#Serialisation
jsonserial= json.dumps(data, indent=2) # serialize 
print(jsonserial)
type(jsonserial)#Data Type is json string
```

```python
jsondeserial = json.loads(jsonserial) # deserialize 
print(jsondeserial)
type(jsondeserial)# Data type is Dictionary
```

```python
#Challenge
#Create a JSON file of the name 'data1.json' which should
#Contain the following data (Serialization)
write_file=open("data_file.json", "w") 
json.dump(data, write_file)

#Load the data from the file 'data1.json' (Deserialization)
readwith open("data_file.json", "r") as read_file:
    data = json.load(read_file)
```
